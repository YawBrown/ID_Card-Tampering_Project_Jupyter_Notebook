ID Card Tampering Detection

This mini project employs Computer Vision to identify the structural similarity with two images. In essence helps distinguish between an original id card and a tampered or fake one.
 

```python
from skimage.metrics import structural_similarity
import imutils
import cv2
from PIL import Image
import requests
```

```python
original=Image.open('id_card_tampering/image/Original.jpg')
tampered=Image.open('id_card_tampering/image/Tampered.jpg')
```

WORK with OPENCV

```python
original=cv2.imread('id_card_tampering/image/original.png')
tampered=cv2.imread('id_card_tampering/image/tampered.png')
```

```python
#Conversion to GrayScale
original_gray=cv2.cvtColor(original,cv2.COLOR_BGR2GRAY)
tampered_gray=cv2.cvtColor(tampered, cv2.COLOR_BGR2GRAY)
```

```python
(score, diff) = structural_similarity(original_gray, tampered_gray, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))
```

```python
#The threshold function of computer vision which applies an adaptive threshold to the image which is stored in the form array. This function transforms the grayscale image into a binary image using a mathematical formula
thresh = cv2.threshold(diff, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)[1]
```

```python
#Diplay difference image with black
print('Different Image')
Image.fromarray(diff)
```

```python
#Display threshold image with white
print('Threshold Image')
Image.fromarray(thresh)
```
